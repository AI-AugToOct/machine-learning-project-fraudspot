```python
import pandas as pd

CSV_PATH = "fake_job_postings.csv" 
```

```python
df = pd.read_csv(CSV_PATH)  
print("Columns BEFORE:", df.columns.tolist())
print("Shape BEFORE:", df.shape)
df.head()
```

```python
df.columns = df.columns.str.strip()
```

### Drop "job_id", "location", "telecommuting", "salary_range" columns

```python
cols_to_drop = ["job_id", "location", "telecommuting", "salary_range"]
df.drop(columns=cols_to_drop, inplace=True, errors="ignore")
```

```python
df.to_csv(CSV_PATH, index=False) 
print(" Changes saved back to file:", CSV_PATH)
```
